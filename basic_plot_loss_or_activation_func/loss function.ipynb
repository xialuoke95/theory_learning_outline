```python
#======================== LTR 损失函数

def ltr_loss(x):
    return np.log(1 + np.exp(-x))

p1 = np.arange(-6, 5, 0.25)
p2 = ltr_loss(p1)
```

```python
plt.figure('Draw') 
plt.plot(p1, p2)  # plot绘制折线图
plt.grid(linestyle='-.')
plt.draw()  # 显示绘图
```
